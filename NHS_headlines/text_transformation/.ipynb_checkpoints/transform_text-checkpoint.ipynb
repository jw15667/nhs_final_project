```python
def split_num_from_str(x):
    pattern = r'(-?[0-9]+\.?[0-9]*)'
    for word in x:
        f = ' '.join(segment for segment in re.split(pattern, x) if segment)
    return f
```

# Text tokenization & stop-words removal

```python
from nltk.corpus import wordnet as wn

def is_proper_noun(tag):
    return tag in ['NNP', 'NNPS']

def is_noun(tag):
    return tag in ['NN', 'NNS']


def is_verb(tag):
    return tag in ['VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ']


def is_adverb(tag):
    return tag in ['RB', 'RBR', 'RBS']


def is_adjective(tag):
    return tag in ['JJ', 'JJR', 'JJS']


def penn_to_wn(tag):
    if is_adjective(tag):
        return wn.ADJ
    elif is_noun(tag):
        return wn.NOUN
    elif is_adverb(tag):
        return wn.ADV
    elif is_verb(tag):
        return wn.VERB
#     elif is_proper_noun(tag):
#         return 'NNP'
    return None
```

```python
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag, word_tokenize
```

```python
def penn_to_wn_pd(tag):
    relevant_words_tokenised = []
    for x in tag:
        if is_adjective(x[1]):
            relevant_words_tokenised.append((x[0], wn.ADJ))
        elif is_noun(x[1]):
            relevant_words_tokenised.append((x[0], wn.NOUN))
        elif is_adverb(x[1]):
            relevant_words_tokenised.append((x[0], wn.ADV))
        elif is_verb(x[1]):
            relevant_words_tokenised.append((x[0], wn.VERB))
    return relevant_words_tokenised
```

# Lemmatization of Words

```python
from nltk.stem import WordNetLemmatizer
lemm = WordNetLemmatizer()
```

```python
def lemm_words(headline):
    lemmed_headline = ''
    for word in headline.split():
        lemm_word = lemm.lemmatize(word)
        lemmed_headline += lemm_word
        lemmed_headline += ' '
    return lemmed_headline
```

```python
data['lemmed'] = data['headline'].apply(lemm_words)
```

```python
def disambiguated_headlines(headline):
    x = disambiguate(headline)
    synsets = []
    for element in x:
        try:
            edit = str(element[1])
            synset = edit.split('Synset(')
            no_brac_syn = synset[1].split(')')
            synsets.append(no_brac_syn[0])
        except:
            element == None
    return synsets
```

```python
def senti_score(word):
    from nltk.corpus import sentiwordnet as swn
    if word != None:
        new_word = word[1:-1]
        x = swn.senti_synset(new_word)
    else:
        x = 'None'
    sentiment = x
    return sentiment

def positive(word):
    from nltk.corpus import sentiwordnet as swn
    from nltk.corpus import sentiwordnet as swn
    if word != None:
        new_word = word[1:-1]
        x = swn.senti_synset(new_word).pos_score()
    else:
        x = 0

    return x


def negative(word):
    from nltk.corpus import sentiwordnet as swn
    from nltk.corpus import sentiwordnet as swn
    if word != None:
        new_word = word[1:-1]
        x = swn.senti_synset(new_word).neg_score()
    else:
        x = 0

    return x
```

```python
def sentiment_anal(hedline):
    from nltk.sentiment.vader import SentimentIntensityAnalyzer 
    sia = SentimentIntensityAnalyzer()
    sentiment = (sia.polarity_scores(hedline)['compound'])
    return sentiment 
```

```python
from datetime import datetime
```

```python
def date_times(x):
    new_date = x[1:-1]
    date = datetime.strptime(new_date, "%d %B %Y").date()        
    return date

def week_num(x):
    week = x.isocalendar()[1]
    return week

def year(x):
    year = x.year
    return year

def week_year(x):
    week = x.isocalendar()[1]
    year = x.year
    unique_key = str(year) + str("-") + str(week)
    return unique_key
```

```python
def findtags(tag_prefix, tagged_text):
    cfd = nltk.ConditionalFreqDist((tag, word) for (word, tag) in tagged_text
                                  if tag.startswith(tag_prefix))
    return dict((tag, cfd[tag].most_common(5)) for tag in cfd.conditions())
```
